```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
img = np.ones((128,128), dtype='uint8')
img *= 255
mask = img.copy()

for i in range(random.randint(1,10)):
    x, y = random.randint(0,128), random.randint(0,128)
    size = random.uniform(0.5,2)
    color = random.randint(0,200)
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(img, 'X',(x,y), font, size, color, 2, cv2.LINE_AA)
    
    
    x, y = random.randint(0,128), random.randint(0,128)
    size = random.uniform(0.5,2)
    color = random.randint(0,200)
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(img, 'A',(x,y), font, size, color, 2, cv2.LINE_AA)
    cv2.putText(mask, 'A',(x,y), font, size, 0, 2, cv2.LINE_AA)
    
fig, ax = plt.subplots(ncols=2)
ax[0].imshow(img,cmap='gray')
ax[1].imshow(mask,cmap='gray')
```
