```python
img = cv2.imread("./PHOTOS_MALARIA_VHIR_UPC/fotos_2015_12_01/P_falciparum/Trofozoits/DSCN0083.JPG",0)
blur = cv2.GaussianBlur(img,(9,9),0)


mode = 2

if mode == 1:
    mode = 'OTSU'
    thresh = filters.threshold_otsu(blur)

else:
    block_size = 261
    offset = 0.0
    mode = 'Adaptative block=%d offset=%.2f' % (block_size, offset)
    thresh = filters.threshold_local(blur, block_size, offset=offset,method='mean')


binary = np.zeros(blur.shape,dtype='uint8')
binary[blur < thresh] = 255

fig, axes = plt.subplots(ncols=2, figsize=(20, 20))
ax0, ax1 = axes

ax0.imshow(binary, cmap='gray')
ax0.set_title(mode)

#ax0.imshow(blur > binary_adaptive)
#ax0.set_title('Adaptive thresholding')
ax1.imshow(exp, cmap='gray')
ax1.set_title('Expected')

for ax in axes:
    ax.axis('off')

plt.show()
```

```python
from scipy import ndimage


eroded = binary.copy()
for t in range(7):
    kernel = np.ones((3, 3), np.uint8)
    eroded = cv2.erode(eroded, kernel)

# find connected components
labeled, nr_objects = ndimage.label(eroded) 
print("Number of objects is {}".format(nr_objects))

print(labeled)

#for cmap in plt.colormaps():
plt.figure(figsize=(20, 10))
plt.imshow(labeled)#CMRmap
plt.show()
```

```python
img = cv2.imread("./PHOTOS_MALARIA_VHIR_UPC/fotos_2015_12_01/P_falciparum/Trofozoits/DSCN0083.JPG",0)
blur = cv2.GaussianBlur(img,(9,9),0)

mode = 2

if mode == 1:
    thresh = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)[1]
else:
    thresh = cv2.adaptiveThreshold(blur,255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV,551,0)

fig, axes = plt.subplots(ncols=2, figsize=(20, 20))
ax0, ax1 = axes

ax0.imshow(thresh, cmap='gray')
ax0.set_title(mode)
ax1.imshow(exp, cmap='gray')
ax1.set_title('Expected')
plt.show()
```

```python
output = cv2.connectedComponentsWithStats(thresh, 4, cv2.CV_32S)
(numLabels, labels, stats, centroids) = output
```

```python


# Map component labels to hue val, 0-179 is the hue range in OpenCV
label_hue = np.uint8(179*labels/np.max(labels))
blank_ch = 255*np.ones_like(label_hue)
labeled_img = cv2.merge([label_hue, blank_ch, blank_ch])

# Converting cvt to BGR
labeled_img = cv2.cvtColor(labeled_img, cv2.COLOR_HSV2BGR)



# set bg label to black
labeled_img[label_hue==0] = 0

fig, axes = plt.subplots(ncols=2, figsize=(20, 20))
ax = axes

ax[0].imshow(cv2.cvtColor(labeled_img, cv2.COLOR_BGR2RGB))
ax[0].title.set_text("Orginal Image")

ax[1].imshow(labels)
ax[1].title.set_text("Orginal Image")
plt.show()
```

```python
labeled_img_RGB = cv2.cvtColor(labeled_img.copy(), cv2.COLOR_BGR2RGB)

shape = labeled_img.shape
labeled_img_r = labeled_img_RGB.reshape(1,shape[0]*shape[1],3)[0]
colors = np.unique(labeled_img_r, axis=0)
colors = np.delete(colors,0,axis=0)
```

```python
def random_color():
    return tuple(np.random.randint(256, size=3))

labeled_img2 = labeled_img_RGB.copy()

for color in colors:
    newColor = random_color()

    red, green, blue = labeled_img2.T 
    selected_color = (red == color[0]) & (green == color[1]) & (blue == color[2])
    labeled_img2[selected_color.T] = newColor

plt.figure(figsize=(10,10))
plt.imshow(labeled_img2)
plt.show()

print(np.max(labeled_img2))
```

```python

marked = cv2.imread(path)
marked_label = (cv2.cvtColor(labeled_img, cv2.COLOR_BGR2RGB)).copy()

for i in range(0, numLabels):
    # extract the connected component statistics and centroid for
    # the current label
    x = stats[i, cv2.CC_STAT_LEFT]
    y = stats[i, cv2.CC_STAT_TOP]
    w = stats[i, cv2.CC_STAT_WIDTH]
    h = stats[i, cv2.CC_STAT_HEIGHT]
    area = stats[i, cv2.CC_STAT_AREA]
    (cX, cY) = centroids[i]
    if area > 1000:
        cv2.circle(marked, (int(cX), int(cY)), 15, (0, 0, 255), -1)
        cv2.circle(marked_label, (int(cX), int(cY)), 15, (255, 255, 255), -1)
    
fig, axes = plt.subplots(ncols=2, figsize=(20, 20))
ax = axes

ax[0].imshow(cv2.cvtColor(marked, cv2.COLOR_BGR2RGB))
ax[1].imshow(marked_label)
plt.show()
```
