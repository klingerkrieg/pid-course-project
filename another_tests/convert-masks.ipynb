```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
import time


plt.rcParams["figure.figsize"] = (5, 5)
# kill axis in rcParams
plt.rc('axes.spines',top=False,bottom=False,left=False,right=False);
plt.rc('axes',facecolor=(1,1,1,0),edgecolor=(1,1,1,0));
plt.rc(('xtick','ytick'),color=(1,1,1,0));
```

```python

def get_image(path, mode=0, size=(640,480), show=False):
    img = cv2.imread(path,mode)
    if mode == 1:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    if img is None:
        print("Não foi possível abrir a imagem:", path)
    res = cv2.resize(img,size)
    return res
```

```python
group = "aglomerado"
gdrive = "."
for i in range(1,16,1):
    path = gdrive+"/samples/%s/blood_smear_%d.JPG" % (group,i)
    print (path)
    if os.path.exists(path):
        img = get_image(path,mode=1)
        plt.imshow(img)
        plt.show()
        
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        
        cv2.imwrite(gdrive+"/samples/normal_gts_font/train/color/%02d.jpg" % (i),img)
```

```python
#copia as labels que ja existem invertendo as cores
gdrive = "."
for i in range(1,16,1):
    path = gdrive+"/samples/normal_gts_font/train/label/%02d.jpg" % i
    if os.path.exists(path):
        img = get_image(path)
        
        inv = np.ones(img.shape)
        inv -= img
        plt.imshow(inv,cmap='gray')
        plt.show()
        
        inv *= 255
        
        cv2.imwrite(gdrive+"/samples/%s_gts_font/train/label-inv/%02d.jpg" % (group,i),inv)
        #plt.imsave(gpath+"/samples/%s_gts_font/gt_%i.png" % (group,i),mask,format='png')
```
